# Hot Words Detection
1. Iterate the csv rows to get the transcripton in column generated_text_ft
2. Detect the hot words : “be careful”, “destroy” and “stranger” in each transcription
3. Save the hot words detected into detected.txt with following columns:
    - filename, generated_text_ft and hot_word 
4. Display the detected.txt file

```python
import pandas as pd

# Load the CSV file
csv_file = '../asr-train/cv-valid-dev_ft.csv'
data = pd.read_csv(csv_file)

# Define hot words
hot_words = ["be careful", "destroy", "stranger"]

# Prepare a list to store filenames with detected hot words
detected_files = []

# Iterate through each row in the CSV
for index, row in data.iterrows():
    transcription = row['generated_text_ft']
    if isinstance(transcription, str):  # Ensure the value is a string
        transcription= transcription.lower()  # convert to lower case
        for word in hot_words:
            if word in transcription:  # Check if any hot word is in the transcription
                detected_files.append({
                    'filename': row['filename'],
                    'generated_text_ft': row['generated_text_ft'],
                    'hot_word': word
                })  # Append the filename, transcription, and hot word
                break  # Stop checking other hot words for this transcription

# Save the detected filenames to a file
detected_file = 'detected.txt'
detected_df = pd.DataFrame(detected_files)
detected_df.to_csv(detected_file, index=False)

# Display the detected files as a table in Jupyter Notebook
display(detected_df)

print(f"Detected filenames with hot words saved to {detected_file}.")
```
